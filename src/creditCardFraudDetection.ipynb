# Spliting the NEW DATASET CREATED INTO FEATURES AND TARGETS

```python
X=new_credit_card_data.drop(columns ='Class', axis=1)
Y=new_credit_card_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

# Split the Data into Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)
#20% of data will be stored in X_test and corresponding class will be stored in Y_test  
#X and Y are the features
#test_size=.2---> means 20% 0f the data will be used for testing 
#stratify=Y--->evenly distributed classes in two sets i.e.train and test set
#random_state=2--->is how do you want to split the data
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training using LOGISTIC REGRESSION 

```python
model = LogisticRegression()
```

```python
#Training the Logistic Regression Model using Training data
model.fit(X_train, Y_train)
#fit--> function helps in fitting the data into our logistic regression model
#X_train-->contains all the features of the trainig data 
#Y_train-->contains the labels of the corresponding training data
```

# Model Evaluation based on Accuracy Score

```python
#finding accuracy score from the training data
#The model will predict the values of Y_train from the X_train values provided then the predicted values will be compared to the original values in the dataset
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training Data", training_data_accuracy)
```

```python
#finding accuracy score from the testing data
#The model will predict the values of Y_test from the X_test values provided then the predicted values will be compared to the original values in the dataset
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on Testing Data", testing_data_accuracy)
print('Accuracy is:',testing_data_accuracy*100)
```
